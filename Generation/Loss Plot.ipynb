```python
# Plotting
import matplotlib
import matplotlib.pyplot as plot
from matplotlib.pyplot import figure
from matplotlib.ticker import MaxNLocator
```

## Model Loss Data

```python
# The loss per epoch (seq_length = 1000, sample_length = 300)
epoch_list = list(range(1, 41))
loss_list = [3.9966, 3.5418, 3.4606, 3.3815, 3.2857, 3.2340, 3.1822, 3.1418, 3.1255, 3.0443, 2.9971, 2.9561, 2.8999, 2.8077, 2.8430, 2.7491, 2.7237, 2.6461, 2.6443, 2.6281, 2.5303, 2.4815, 2.4892, 2.3747, 2.2870, 2.3215, 2.2654, 2.2489, 2.0884, 2.0694, 2.1884, 1.9790, 2.0509, 1.9381, 1.9074, 1.9020, 1.8158, 1.8269, 1.7918, 1.7771]
```

## Plot

```python
# Set the plot font
font = {'family' : 'DejaVu Sans',
        'size'   : 18}

matplotlib.rc('font', **font)



# Set the plot size
figure(figsize=(8, 6), dpi=80)

# Plot the data
plot.plot(epoch_list, loss_list, marker='.', markersize=15)

# Set the titles
plot.title('RNN Loss per Epoch')
plot.xlabel('Epoch')
plot.ylabel('Loss')

# Force the x-axis to display integer ticks
ax = plot.gca()
ax.xaxis.set_major_locator(MaxNLocator(integer=True))

plot.show()
```
