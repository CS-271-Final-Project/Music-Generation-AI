## Add Pitches to an Array

```python
def add_pitches(file_name, pitch_array):
    # Open the csv file
    with open(file_name, 'r') as csvfile:
        datareader = csv.reader(csvfile)
        
        # Iterate over each row
        for row in datareader:
            if row:
                for pitch in row:
                    pitch_array[int(pitch)] += 1
```

## Count the Frequency of each Pitch

```python
# An array containing the frequency of each midi pitch
pitch_array = np.zeros(128)

# Go to the corpus directory
os.chdir(corpus_path)

# Count the files processed
curr_file = 0

# Add the notes of each score to the graph
for song_file in glob('*.csv'):
    curr_file += 1
    
    add_pitches(song_file, pitch_array)
    
    if curr_file % 100 == 0:
        print(str(curr_file))

# Plot the graph
plt.bar(list(range(128)), pitch_array)
plt.show()
```
