```python
# Data Processing
import numpy as np
from numpy import genfromtxt
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Plotting
import matplotlib
import matplotlib.pyplot as plot
from matplotlib.pyplot import figure
```

## Settings

```python
classical_features_path = 'C:\\Users\\alext\\Desktop\\School\\2021 Spring\\CS 271\\Final Project\\Feature Extraction\\Classical\\Classical Features.csv'
jazz_features_path = 'C:\\Users\\alext\\Desktop\\School\\2021 Spring\\CS 271\\Final Project\\Feature Extraction\\Jazz\\Jazz Features.csv'

# Set the plot font
font = {'family' : 'DejaVu Sans',
        'size'   : 18}

matplotlib.rc('font', **font)
```

## Normalize Feature Vectors

```python
def normalize_features(features):
    # Compute the mean and varience of each feature
    scaler = StandardScaler().fit(features)
    # Use the computed values to normalize the features
    normalized_features = scaler.transform(features)
    
    return normalized_features
```

## Plot All Feature

```python
def plot_all_features(group1, group2, num_experiments):
    # Assume all experiments contain the same number of features
    num_features = len(group1[0])
    
    # Set the plot size
    figure(figsize=(8, 6), dpi=80)
    
    # Plot each feature of each sample
    for feature in range(num_features):
        for experiment in range(num_experiments):
            plot.scatter(feature, group1[experiment][feature], color='red', alpha=0.5)
            plot.scatter(feature, group2[experiment][feature], color='blue', alpha=0.5)
    
    # Set the plot title
    plot.title('All Features 0-{}'.format(num_features))
    
    # Hide the y-azis lables
    ax = plot.gca()
    ax.axes.yaxis.set_visible(False)
    
    # set the range of the y axis
    plot.ylim([-9, 9])
    
    # Display the plot
    plot.show()
```

## Visualize the Data

```python
# Load the data set features
classical_features = genfromtxt(classical_features_path, delimiter=',')
jazz_features = genfromtxt(jazz_features_path, delimiter=',')

# Normalize the features
classical_normalized = normalize_features(classical_features)
jazz_normalized = normalize_features(jazz_features) 

plot_all_features(classical_normalized, jazz_normalized, 5)
```
